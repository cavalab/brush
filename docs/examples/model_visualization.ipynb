```python
import pandas as pd
from brush import BrushRegressor

# load data
df = pd.read_csv('./datasets/d_enc.csv')
X = df.drop(columns='label')
y = df['label']

# import and make a regressor
est = BrushRegressor()

# use like you would a sklearn regressor
est.fit(X,y)
y_pred = est.predict(X)

print('score:', est.score(X,y))
```

```python
import graphviz
from brush import SearchSpace, Dataset


model = est.best_estimator_.prg.get_dot_model()
g = graphviz.Source(model)
g
```
